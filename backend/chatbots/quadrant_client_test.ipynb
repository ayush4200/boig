```python
from qdrant_client import QdrantClient
import os
from langchain.vectorstores import Qdrant
from langchain.embeddings import OpenAIEmbeddings

from dotenv import load_dotenv
```

```python
url = os.environ.get("QUADRANT_URL")
api_key = os.environ.get("QUADRANT_API_KEY")
openai_api_key = os.environ.get("OPENAI_API_KEY")
```

```python
load_dotenv()
```

```python
embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)
```

```python
client = QdrantClient(
    url=url,
    api_key=api_key,
    prefer_grpc=True
)
```

```python
# collection_info = client.get_collection(collection_name="test-agent-1")
```

```python
qdrant = Qdrant(client=client, collection_name="test-agent-1", embeddings=embeddings)
```

```python
qdrant
```

```python
# query = "BMW iDrive with Operating System 8 is gradually launched in all BMW vehicle classes. The technology will be used for the first time in the BMW iX"
query = "Is bmw idrive launched in all classes?"
found_docs = qdrant.similarity_search_with_score(query, k=1)
```
